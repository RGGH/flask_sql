## convert to tokens

```python
word_tokens = word_tokenize(str(tx[1]))

for w in (word_tokens):
    print(w)
```

## POS Tagging (Parts of Speech Tagging)  also called grammatical tagging

```python
tag1 = pos_tag(word_tokens)
print(tag1)
```

```python
for w in word_tokens:
    print(ps.stem(w))   # Passing word tokens into stem method of Porter Stemmer
```
